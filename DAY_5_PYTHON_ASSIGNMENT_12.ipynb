```python
'''3. Write a menu driven program to practice List functions. Validate input data wherever
required.
Display following menu:
1. Accept Data
a) add at last position
b) add at given position
2. Delete data by value
 display message deleted successfully
or not found
3. delete data by position
a) delete last element
b) delete from particular position
4. sort
a) ascending
b) descending
5. reverse
6. Print in sorted order without changing original list
7. print in reverse order without changing original list
8. display data
a) normal
b) numbered'''


list1 = [12, 3, 5, 14, 7, 9, 6, 4]
user_input = 0
while user_input != 9:
    print("1. Accept Data")
    print("2. Delete data by value")
    print("3. Delete data by position")
    print("4. Sort")
    print("5. Reverse")
    print("6. Print in sorted order without changing original list")
    print("7. Print in reverse order without changing original list")
    print("8. Display data")
    print("9. exit")
    user_input = input("Choose an option: ")
    if user_input == "1":
        print("a) Add at last position")
        print("b) Add at given position")
        choice = input("choose from a or b: ")
        if choice == "a":
            num = int(input("Enter a number to add at last position: "))
            list1.append(num)
            print("Successfully added at last position\n")
            print(list1)
        elif choice == "b":
            num, position = map(int, input("Enter a number and position seperated by space: ").split())
            list1.insert(position, num)
            print("Successfully added at given position\n")
            print(list1)
        else:
            print("wrong input")    
    elif user_input == "2":
        value = int(input("Enter a value you want to delete: "))
        if value in list1:
            list1.remove(value)
            print("removed successfully\n")
            print(list1)
        else:
            print("value not found")
    elif user_input == "3":
        print("a) Delete last element")
        print("b) Delete from particular position")
        choice = input("choose from a or b: ")
        if choice == "a":
            list1.pop()
            print("last element deleted\n")
            print(list1)
        elif choice == "b":
            position = int(input(f"choose a position from 0-{len(list1)-1} to delete data: "))
            list1.pop(position)
            print(f"value deleted from the {position} position\n")
            print(list1)
        else:
            print("wrong input")
    elif user_input == "4":
        print("a) Ascending")
        print("b) Descending")
        choice = input("choose from a or b: ")
        if choice == "a":
            ls=sorted(list1)
            print("List sorted in ascending order is: \n",ls)
            print(ls)
        elif choice == "b":
            lsr=sorted(list1, reverse=True)
            print("List sorted in descending order is: \n",lsr)
            print(lsr)
        else:
            print("wrong input")
    elif user_input == "5":
        list1 = list(reversed(list1))
        print("list reversed successfully:\n")
        print(list1)
    elif user_input == "6":
        print("List in sorted order without changing original list:\n")
        print(sorted(list1))
        
    elif user_input == "7":
        print("List in reverse order without changing original list:\n")
        print(list(reversed(list1)))
    elif user_input == "8":
        print("a) Normal")
        print("b) Numbered")
        choice = input("choose from a or b: ")
        if choice == "a":
            print("List in normal mode:\n")
            print(list1)
        elif choice == "b":
            print("List in numbered mode:\n")
            for x, y in enumerate(list1, 1):
                print(x,"------>",y)
        else:
            print("WRONG INPUT !!!")
    elif user_input =="9":
        break
```

```python
'''6. Write a menu driven program to practice Set functions.
Write a program to accept names from users and store it in a set.
Display the following menu:
print("""1. delete element if exists otherwise
 do not show any errr""")
print("2. add a elemet")
print("3. create one more set")
print("4. union of 2 sets")
print("4. intersection of 2 sets")
print("5. difference of 2 sets")
print("6. convert set into frozenset")
print("6. exit") '''


setA = set(map(str, input("Enter names seperated by a space: ").split()))
user_input = "0"
while user_input != "8":
    print("1. Enter an element you want to delete: ")
    print("2. add an element to set")
    print("3. create a set")
    print("4. union of 2 sets")
    print("5. intersection on 2 set")
    print("6. diffrence of 2 sets")
    print("7. convert set into frozen set")
    print("8. exit")
    user_input = input("choose an option: ")
    if user_input == "1":
        name = input("Enter a name to delete: ")
        print("Original set: ",setA)
        try:
            setA.remove(name)
            print("Element deleted successfully from the set\n")
            print(setA)
        except KeyError:
            pass
        except AttributeError:
            print("This set is a frozenset, delete is not allowed\n")
    elif user_input == "2":
        print("Original set: ",setA)
        try:
            setA.add(input("Enter a name to add: "))
            print("Element added to the set\n")
            print(setA)
        except AttributeError:
            print("This set is a frozenset, element insertion is not allowed\n")
    elif user_input == "3":
        setB = set(map(str,input("Enter the names space seperated to add to new set: ").split()))
        print("New set created successfully\n")
        print(setB)
    elif user_input == "4":
        print("Original set: ",setA)
        try:
            if setB:
                print("second set: ",setB)
                print("Union of the two set's is:\n")
                print(setA | setB)               
        except NameError:
            print("Create a new set to find union by selecting option 3 \n")
    elif user_input == "5":
        print("Original set: ",setA)
        try:
            if setB:
                print("second set: ",setB)
                print("Intersection of the two set's is:\n")
                print(setA & setB)
        except NameError:
            print("Create a new set to find intersection by selecting option 3 \n")
    elif user_input == "6":
        print("Original set: ",setA)
        try:
            if setB:
                print("second set: ",setB)
                print("Difference of the two set's is:\n")
                print(setA - setB)
        except NameError:
            print("create a new set to find difference by selecting option 3 \n")
    elif user_input == "7":
        setA = frozenset(setA)
        print("Set is converted into frozenset successfully\n")
        print(setA)
    elif user_input == "8":
        break
    else:
        print("Choose an option from these: ")
```
