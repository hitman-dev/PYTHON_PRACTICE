```python
'''Q2. Given a dictionary of students and their favourite colours:
people={'Arham':'Blue','Lisa':'Yellow',''Vinod:'Purple','Jenny':'Pink'}
1. Find out how many students are in the list
2. Change Lisa’s favourite colour
3. Remove 'Jenny' and her favourite colour
4. Sort and print students and their favourite colours alphabetically by name'''


choice = 0
people={'Arham':'Blue','Lisa':'Yellow','Vinod':'Purple','Jenny':'Pink'}
print('''1. Find out how many students are in the list
2. Change Lisa’s favourite colour
3. Remove 'Jenny' and her favourite colour
4. Sort and print students and their favourite colours alphabetically by name''')
while choice != 5:
    choice = int(input("enter your choice:"))
    if choice == 1:
        print(f"The are {len(people)} students in the list")
    elif choice == 2:
        print("Are you sure to change Lisa's favorite colour ")
        ans = input("Y/N:")
        if ans=='Y':
            newfavCol=input("enter new favourite colours of Lisa:")
            people['Lisa'] = newfavCol
            print("Lisa's favorite color changed to ",newfavCol)
            print(people)
    elif choice == 3:
        people.pop('Jenny')
        print("Removed Jenny and her favorite color")
        print(people)
    elif choice == 4:
        for i,k in sorted(people.items()):
            print(i, ":", k)
```
