```python
'''1.use -wheather_data_cities.csv
draw a bar graph and pie chart
on x axis use city names
and on y axis use max of temperature
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

weatherdf = pd.read_csv('weather_by_cities.csv')
weatherdf.head()
```

```python
dftemp = weatherdf.groupby(["city"])["temperature"].max()
dftemp
```

```python
dftemp.plot(kind='bar', title='Max temp by cities', ylabel='Max Temp',xlabel='Cities')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(8,6))
plt.pie(dftemp,autopct='%.2f %%', labels=dftemp.index,textprops=dict(color="#00FFFF"))
plt.title('Max temp by cities', fontsize=18,color='#00FFFF')
plt.legend(dftemp.index,  title="Cities",loc='right', bbox_to_anchor=(1, 1,0.5,0))
plt.show()
```

```python
'''2.
#create a list for storing year 2010 to 2014
#create a list for each year for storing sales amount for 5 products in each years
#draw pie chart and stack graph to compare sales
1. Yearly comparison
Year and max sale in each year
2. Draw separate graph for each year(5 different pie charts)
Product and average sale of that product'''



```

```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
year=[]
sales={}
y=2010
for i in range(0,5):
    year.append(y)
    sales = { year : []  for year in year }
    y+=1

print(year)
print(sales)

for i in sales:
    for j in range (0,5):
        sales[i].append(random.randint(5000, 50000))
print(sales)
```

```python
products=['TV','AC','Washing machine','Fridge','Fan']
dfsh = pd.DataFrame(sales,index=products)
dfsh
```

```python

dfs=dfsh.sum(axis=0)
 
print(dfs)
```

```python

dfs.plot(kind='bar', title='Max sales', ylabel='Year wise Max sales',xlabel='year')
plt.xticks(rotation=45)
```

```python
# 1. Yearly comparison
# Year and max sale in each year
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format
plt.figure(figsize=(8,6))
plt.pie(dfsh.sum(axis=0),autopct='%.2f %%', labels=dfsh.sum(axis=0),textprops=dict(color="#00FFFF"))
plt.legend(dfsh.columns,  title="Years",loc='right', bbox_to_anchor=(1, 1,0.5,-0.5))
plt.title('Year and max sale in each year ', fontsize=18,color='#00FFFF')
plt.show()
```

```python
# 2. Draw separate graph for each year(5 different pie charts)
# Product and average sale of that product'
#Pie chart for each year

import matplotlib.pyplot as plt
for i in range(len(dfsh.index)):
    fig, ax = plt.subplots(1,1)
    fig.set_size_inches(5,5)
    dfsh.iloc[i].plot(kind='pie', ax=ax, autopct='%1.1f%%',labels=dfsh.index,textprops=dict(color="#00FFFF"))
    plt.legend(dfsh.index,  title="Products",loc='right', bbox_to_anchor=(1, 1,0.5,-0.5))
    plt.title(f'Average sale of each product in {year[i]}', fontsize=18,color='#00FFFF')
    plt.show()
```

```python
#Stackplot graph
plt.figure(figsize=(10,8))
yearList = ['2010','2011','2012','2013','2014']
plt.stackplot(yearList,dfsh.values.T)
plt.legend(labels=dfsh.index)
plt.xlabel('Year')
plt.ylabel('Sale')
plt.title('Sales of each produt by year')
plt.show()
```
