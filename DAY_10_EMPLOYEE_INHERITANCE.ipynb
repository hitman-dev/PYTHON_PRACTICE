```python
from abc import ABC, abstractmethod
object
class Person:
    def __init__(self, eid=0, nm=""):
        self.__eid = eid
        self.__pname = nm
   
    def set_eid(self, eid):
        self.__eid = eid
    def set_pname(self, nm):
        self.__pname = nm
    def get_pname(self):
        return self.__pname
    def get_eid(self):
        return self.__eid
    def __str__(self):
        return f"E.Id : {self.__eid}, Name : {self.__pname}"


class employee(Person, ABC):
    def __init__(self, eid=0, nm="",dept="",desg="") :
        super().__init__(eid, nm)
        self.__dept = dept
        self.__desg = desg
    @abstractmethod
    def calculatesal(self):
        pass
    def __str__(self):
        return super().__str__()+f"Department : {self.__dept} , Designation: {self.__desg}"


class salariedemp(employee):
    def __init__(self, eid=0, nm="",dept="",desg="",sal=0):
        super().__init__(eid, nm,dept,desg)
        self.__sal = sal
        self.__bonus = 0.2*sal
    def set_sal(self, sal):
        self.__sal = sal
    def get_sal(self):
        return self.__sal

    def __str__(self):
        return super().__str__()+f" Salory: {self.__sal} Rs, Bonus : {self.__bonus} Rs"

    def calculatesal(self):
        return f"Net salory={round(self.__sal+(0.1*self.__sal)+(0.15*self.__sal)-(0.08*self.__sal))} Rs"


class contractemp(employee):
    def __init__(self,eid=0, nm="",dept="",desg="",hrs_worked="" , hourly_charges=""):
        super().__init__(eid, nm,dept,desg)
        self.__hrs_worked = hrs_worked
        self.__hourly_charges = hourly_charges

    def set_hourly_charges(self, hourly_charges):
        self.__hourly_charges = hourly_charges     
    def get_hourly_charges(self):
        return self.__hourly_charges
    def set_hrs_worked(self, hrs_worked):
        self.__hrs_worked = hrs_worked    
    def get_hrs_worked(self):
        return self.__hrs_worked
        

    def __str__(self):
        return super().__str__()+f" Hours worked: {self.__hrs_worked} hrs, Hourly charges: {self.__hourly_charges} Rs"

    def calculatesal(self):
        return f"Net salory={round(self.__hrs_worked*self.__hourly_charges)} Rs"

# if __name__ == "__main__":
#     p1 = salariedemp(12,'kishori','training','Manager',1111)
#     p2 = contractemp(12,'esha','insurance','Manager',12,1000.00)
#     print(p1)
#     print(p2)
#     print(p1.calculatesal())
#     print(p2.calculatesal())
#     salariedemp.sal=100
#     print(p1)
```

```python
lst = []
emp1 = salariedemp(12,'kishori','training','Manager',1111)
emp2 = salariedemp(13,'rajan','game','game designer',11111)
emp3 = contractemp(14,'esha','insurance','Manager',12,1000.00)
emp4 = contractemp(15,'revati','ux','ux designer',111,1500.00)
lst.append(emp1)
lst.append(emp2)
lst.append(emp3)
lst.append(emp4)

def add_NewEmploy(ch):
    eid = int(input("Enetr new Employee id"))
    nm = input("Enetr new employee name")
    dept = input("Enetr new Employee Department")
    desg = input("Enetr new Employee Designation")
    if ch == 1:
        sal = int(input("Enter new Employee salary(CTC)"))
        e = salariedemp(eid, nm,dept,desg,sal)
        lst.append(e)
    elif ch == 2:
        hrs_worked=int(input("Enetr new Employee worked hrs"))
        hourly_charges=int(input("Enetr new Employee hourly charges"))
        e = contractemp(eid, nm,dept,desg,hrs_worked,hourly_charges)
        lst.append(e)
    else:
        return False
        

def searchById(eid):
    for s in lst:
        if s.get_eid()==eid:
            return s
    else:
        return None

def deleteById(eid):
        e=searchById(eid)
        if e!=None:
            print(e)
            ch=input("Do you want to delete this employee data y/n:")
            if ch=="y":
                lst.remove(e)
                print(f"Employee with ID {e} deleted successfully")
            else:
                return False
        else:
            return False

def modiefysal(ch):
        if ch == 1:
            eid = int(input("Enter Employee id whose salary you want to modify:"))
            e=searchById(eid)
            if e!=None:
                print(e)
                m=e.get_pname()
                ch=input(f"Do you want to modiefy {m} salory y/n:-")
                if ch=="y":
                    new=int(input(f"Enter new salary for {m}:-"))
                    old=e.get_sal()
                    name1=e.set_pname(new)
                    print(f"Employee with ID {eid} and name {m},salory modiefied from old salory of {old} to {new}")
                else:
                    return False
            else:
                return False
        elif ch == 2:
            eid = int(input("Enter Employee id whose salary you want to modify:"))
            e=searchById(eid)
            if e!=None:
                print(e)
                m=e.get_pname()
                ch=input(f"Do you want to modiefy {m} salory y/n:-")
                if ch=="y":
                    print("1. Modiefy worked hours\n 2. Modiefy hourely charges")
                    opn = int(input("Enetr your choice"))
                    if opn==1:
                        newwh=int(input(f"Enter new worked hours data for {m}:-"))
                        oldwh=e.get_hrs_worked()
                        nwh=e.set_hrs_worked(newwh)
                        print(f"Employee with ID {eid} and name {m},worked hours modiefied from {oldwh}hrs to {newwh}hrs")
                    elif opn==2:
                        newhc=int(input(f"Enter new hourly charges for {m}:-"))
                        oldhc=e.get_hourly_charges()
                        nhc=e.set_hourly_charges(newhc)
                        print(f"Employee with ID {eid} and name {m},hourly charges modiefied from {oldhc} Rs to {newhc} Rs")
                    else:
                        return False
                else:
                    return False
            else:
                return False
        else:
            return False


def display_All():
    for e in lst:
        print(e)


def calculate_salary(eid):
    s = searchById(eid)
    if s != None:
        return s.calculatesal()  # dynamic polymorphism
    else:
        return None
```
